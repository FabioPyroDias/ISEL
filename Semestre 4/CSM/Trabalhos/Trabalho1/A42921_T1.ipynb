# 6
Construa uma função que realize o algoritmo de dithering Floyd Steinberg. Esta função recebe uma matrix (com
os pixeis em tons de cinzento) e devolve uma matrix com valores a preto e branco. Este algoritmo aproxima
cada pixel da imagem (x) ao valor mais próximo (preto ou branco) e o erro é difundido para os pixeis adjacentes
seguindo o método:<p>

$${\displaystyle {\begin{bmatrix}&&*&{\frac {\displaystyle 7}{\displaystyle 16}}&\ldots \\\ldots &{\frac {\displaystyle 3}{\displaystyle 16}}&{\frac {\displaystyle 5}{\displaystyle 16}}&{\frac {\displaystyle 1}{\displaystyle 16}}&\ldots \\\end{bmatrix}}}$$

Aplique esta função à imagem em tons de cinzento que está na variável x_img_g.

```python
def dither(img):
    newImg = img.copy()
    
    for pixelY in range(len(img)):
        for pixelX in range(len(img[pixelY])):
            oldpixel = newImg[pixelX][pixelY]
            newpixel = 255 if oldpixel > 128 else 0
            newImg[pixelX][pixelY] = newpixel
            quant_error = oldpixel - newpixel
            if(pixelX != len(img) - 1):
                newImg[pixelX + 1][pixelY] = newImg[pixelX + 1][pixelY] + quant_error * 7 / 16
                
                if(pixelY != len(img[pixelX]) - 1):
                    newImg[pixelX + 1][pixelY + 1] = newImg[pixelX + 1][pixelY + 1] + quant_error * 1 / 16
            
            if(pixelY != len(img[pixelX]) - 1):
                newImg[pixelX - 1][pixelY + 1] = newImg[pixelX - 1][pixelY + 1] + quant_error * 3 / 16
                newImg[pixelX][pixelY + 1] = newImg[pixelX][pixelY + 1] + quant_error * 5 / 16
            
    return newImg

y = dither(x_img_g)
plt.figure(figsize = (14, 8))
plt.imshow(y,cmap='gray', vmin=0, vmax=255)
```

# 8
Crie uma função que apresente uma imagem com dimensão de 100 × 100 pixeis como se apresenta na figura. O
ângulo de cada sector é dado por parâmetro passado à função (o ângulo é um valor inteiro entre 0 e 360 graus).

```python
def radial(angulo):
    
    imagem = np.zeros((200, 200))
    x, y = np.meshgrid(np.arange(-100, 100), np.arange(-100, 100))
    z = x + y*1j    
    
    #Como o array vem todo em 0's, basta mudar o que é 255.
    #Arranjo uma matriz de números complexos e arranjo o ângulo.
    #Este vem entre -180 e 180, logo, tem de ser mapeado para 0 e 360.
    #Depois, fazemos a divisão do ângulo do pixel para o angulo passado.
    #Aproxima-mos para o menor inteiro.
    #Fazemos esta operação toda e procuramos a qual dos "sectores" pertence.
    #Se for par: Fica com o valor 255 (branco)
    #Se não for par: Este não é alterado e fica com o 0 (preto) que já tinha antes
    
    imagem[(np.floor(((360 + np.angle(z, deg=True)) % 360) / angulo) % 2 == 0)] = 255
    
    return imagem
    
a = radial(150)
plt.figure(figsize=(7, 7))
plt.imshow(a, cmap="gray")
```
