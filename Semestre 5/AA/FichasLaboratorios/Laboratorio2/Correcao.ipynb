```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
D=pickle.load(open('Lab2-data.p','rb'))
print(D.keys())
```

P.1.1.

```python
X = D.data
y = D.target
f1 = D.foldTrain
f2 = D.foldTest
X1 = X[:, f1]
y1 = y[f1]
X2 = X[:, f2]
y2 = y[f2]
```

```python
np.unique(y2)
```

```python
m0 = np.mean(X1[:, y1==0], axis=1)
m1 = np.mean(X1[:, y1==1], axis=1)
m2 = np.mean(X1[:, y1==2], axis=1)
m3 = np.mean(X1[:, y1==3], axis=1)

print(m0)
print(m1)
print(m2)
print(m3)
```

```python
plt.plot(X[0, y==0], X[1, y==0], '.')
plt.plot(X[0, y==1], X[1, y==1], '.')
plt.plot(X[0, y==2], X[1, y==2], '.')
plt.plot(X[0, y==3], X[1, y==3], '.')
plt.axis('scaled')
plt.grid(True)
```

P.1.2.

```python
Xa = X2 - m0[:, np.newaxis]
Xb = X2 - m1[:, np.newaxis]
Xc = X2 - m2[:, np.newaxis]
Xd = X2 - m3[:, np.newaxis]

d0 = np.sqrt(np.sum(Xa**2, axis=0))
d1 = np.sqrt(np.sum(Xb**2, axis=0))
d2 = np.sqrt(np.sum(Xc**2, axis=0))
d3 = np.sqrt(np.sum(Xd**2, axis=0))

D = np.vstack((d0, d1, d2, d3))
y2e = np.argmin(D, axis=0)

print(confusion_matrix(y2, y2e))
np.sum(y2 != y2e)
```

```python
plt.plot(X[0, y==0], X[1, y==0], '.')
plt.plot(X[0, y==1], X[1, y==1], '.')
plt.plot(X[0, y==2], X[1, y==2], '.')
plt.plot(X[0, y==3], X[1, y==3], '.')
plt.plot(m0[0], m0[1], '.')
plt.plot(m1[0], m1[1], '.')
plt.plot(m2[0], m2[1], '.')
plt.plot(m3[0], m3[1], '.')
plt.axis('scaled')
plt.grid(True)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X1.T, y1)
y2e = knn.predict(X2.T)
print(confusion_matrix(y2, y2e))
print(np.sum(y2 != y2e))

plt.plot(X2[0, y2==0], X2[1, y2==0], '.')
plt.plot(X2[0, y2==1], X2[1, y2==1], '.')
plt.plot(X2[0, y2==2], X2[1, y2==2], '.')
plt.plot(X2[0, y2==3], X2[1, y2==3], '.')
plt.plot(X2[0, y2!= y2e], X2[1, y2!=y2e], 'ok', alpha=0.2)
plt.axis('scaled')
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X1.T, y1)
y2e = knn.predict(X2.T)
print(confusion_matrix(y2, y2e))
print(np.sum(y2 != y2e))

plt.plot(X2[0, y2==0], X2[1, y2==0], '.')
plt.plot(X2[0, y2==1], X2[1, y2==1], '.')
plt.plot(X2[0, y2==2], X2[1, y2==2], '.')
plt.plot(X2[0, y2==3], X2[1, y2==3], '.')
plt.plot(X2[0, y2!= y2e], X2[1, y2!=y2e], 'ok', alpha=0.2)
plt.axis('scaled')
```
