## Discriminantes Logísticos

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegressionticRegression
import pickle
```

### Classificação Binária: Dados Sintéticos binClassData

```python
D = pickle.load(open('PreparacaoLaboratorio5/binClassData.p', 'rb'))
```

```python
D.keys()
```

```python
X1 = D['Xtrain']
y1 = D['trueClassTrain']
X2 = D['Xtest']
y2 = D['trueClassTest']
```

```python
plt.plot(X1[0, y1==0], X1[1, y1==0], 'sb', X1[0, y1==1], X1[1, y1==1], 'sr', alpha=0.4)
plt.plot(X2[0, y2==0], X2[1, y2==0], '*b', X2[0, y2==1], X2[1, y2==1], '*r', ms=10)
plt.axis('scaled')
plt.grid(True)
```

Vamos agora treinar um determinante logístico, com diferentes tipos de regularização, l1 e l2, e também com diferentes pesos de regularização.

```python
dl = LogisticRegression(penalty='l2', C=0.1, random_state=42, solver='saga').fit(X1.T, y1)
```

Neste caso estamos a inicializar o gerador de números aleatórios, random_state, pois estes algoritmos têm de encontrar os pesos iterativamente e, na primeira iteração, já têm de ter um valor. Se forçarmos o valor a ser sempre o mesmo, 42 neste caso, podemos compará-lo com outras mudanças feitas sem termos de considerar a hipótese que os valores da primeira iteração foram mudados.

Também usamos o "solver=saga" porque este dá tanto para a regularização l1 como para a l2. Nem todos os solvers são capazes de fazer um tipo ou outro tipo de regularização. Uns só funcionam para o l1 e outros para o l2.

```python
print('Train Score: ', dl.score(X1.T, y1))
print('Test Score: ', dl.score(X2.T, y2))
```

Como já podiamos inferir pelo plot, existem dados sobrepostos, logo, o resultado ia ser sempre diferente de 100%. Caso fosse 100%, era mau sinal pois indicava sobreaprendizagem.<br>
Entretanto, com o score, confirmamos isso.

Vamos observar onde é que o classificador inseriu a fronteira de decisão para estes dados.

```python
w = dl.coef_
w = w.squeeze()

w0 = dl.intercept_

print('w0: ', w0)
print('w: ', w)
```

A razão de usarmos o "squeeze" é para eliminarmos a dimensão extra nos w's.

Já temos os pesos óptimos para a classificação dos pontos.<br>
Com estes, podemos usar a fórmula : x2 = (-1/w2)*(w0, w1 * x1)<br>

Ou seja, se pegarmos em dois pontos de x1 e obtemos o x2, tal como uma função linear y = mx + b

```python
x1 = np.array([-2, 2])
x2 = -1 /w[1] * (w0 + w[0] * x1)

print('x1: ', x1)
print('x2: ', x2)

plt.plot(X1[0, y1==0], X1[1, y1==0], 'sb', X1[0, y1==1], X1[1, y1==1], 'sr', alpha=0.4)
plt.plot(X2[0, y2==0], X2[1, y2==0], '*b', X2[0, y2==1], X2[1, y2==1], '*r', ms=10)
plt.plot(x1, x2, '--k')
plt.axis('scaled')
plt.axis([-2, 2, -1.5, 1.5])
plt.grid(True)
```

Dado que o nosso w2 (w[1]) era muito pequeno, na fórmula estamos a dividir 1 por esse w2. Ou seja, o valor de x2 irão ser extremamente elevados.<br>
Dai estarmos a restringir a vertical entre -1.5 e 1.5.

Vamos agora repetir o mesmo processo, mas usando o l1 em vez do l2.

```python
dl = LogisticRegression(penalty='l1', C=0.1, random_state=42, solver='saga').fit(X1.T, y1)

print('Train Score: ', dl.score(X1.T, y1))
print('Test Score: ', dl.score(X2.T, y2))

w = dl.coef_
w = w.squeeze()

w0 = dl.intercept_

print('w0: ', w0)
print('w: ', w)

x1 = np.array([-2, 2])
x2 = -1 /w[1] * (w0 + w[0] * x1)

print('x1: ', x1)
print('x2: ', x2)

plt.plot(X1[0, y1==0], X1[1, y1==0], 'sb', X1[0, y1==1], X1[1, y1==1], 'sr', alpha=0.4)
plt.plot(X2[0, y2==0], X2[1, y2==0], '*b', X2[0, y2==1], X2[1, y2==1], '*r', ms=10)
plt.plot(x1, x2, '--k')
plt.axis('scaled')
plt.axis([-2, 2, -1.5, 1.5])
plt.grid(True)
```

Algo que podemos observar logo é a descrepância entre o score dos dados de treino e dos dados de teste.<br>
Isto pode ser uma indicação que a regularização l2 pode ser mais indicada para este tipo de dados.

Também observamos que w2 é 0. Logo, na fórmula, o x2 vai dar infinito. (1/0)<br>
E temos essa confirmação no print dos valores de x2.

Neste caso, a fronteira é completamente na vertical, apesar do plot não ter sido capaz de o representar.

Vamos agora repetir estes dois processos, mas com uma regularização bastante mais leve. Ou seja, aumentar o valor de C.

```python
dl = LogisticRegression(penalty='l2', C=100, random_state=42, solver='saga').fit(X1.T, y1)

print('Train Score: ', dl.score(X1.T, y1))
print('Test Score: ', dl.score(X2.T, y2))

w = dl.coef_
w = w.squeeze()

w0 = dl.intercept_

print('w0: ', w0)
print('w: ', w)

x1 = np.array([-2, 2])
x2 = -1 /w[1] * (w0 + w[0] * x1)

print('x1: ', x1)
print('x2: ', x2)

plt.plot(X1[0, y1==0], X1[1, y1==0], 'sb', X1[0, y1==1], X1[1, y1==1], 'sr', alpha=0.4)
plt.plot(X2[0, y2==0], X2[1, y2==0], '*b', X2[0, y2==1], X2[1, y2==1], '*r', ms=10)
plt.plot(x1, x2, '--k')
plt.axis('scaled')
plt.axis([-2, 2, -1.5, 1.5])
plt.grid(True)
```

As primeiras coisas que nos saltam à vista são as melhorias dos scores. Não só estã mais elevados individualmente, como a descrepância entre eles é muito mais pequena.

Também podemos observar que os pesos de w são agora mais elevados, o que vai influenciar a inclinação da reta.

Isto foi para a regularização l2. Vamos observar o caso da l1.

```python
dl = LogisticRegression(penalty='l1', C=100, random_state=42, solver='saga').fit(X1.T, y1)

print('Train Score: ', dl.score(X1.T, y1))
print('Test Score: ', dl.score(X2.T, y2))

w = dl.coef_
w = w.squeeze()

w0 = dl.intercept_

print('w0: ', w0)
print('w: ', w)

x1 = np.array([-2, 2])
x2 = -1 /w[1] * (w0 + w[0] * x1)

print('x1: ', x1)
print('x2: ', x2)

plt.plot(X1[0, y1==0], X1[1, y1==0], 'sb', X1[0, y1==1], X1[1, y1==1], 'sr', alpha=0.4)
plt.plot(X2[0, y2==0], X2[1, y2==0], '*b', X2[0, y2==1], X2[1, y2==1], '*r', ms=10)
plt.plot(x1, x2, '--k')
plt.axis('scaled')
plt.axis([-2, 2, -1.5, 1.5])
plt.grid(True)
```

Continuamos a ter muito bons scores com pouca descrepância.<br>
Os pesos continuam altos, assim como existe uma boa inclinação da reta.

Também existe a opção de não usar regularização. Vamos observar os resultados nesse caso.

```python
dl = LogisticRegression(penalty='none', C=100, random_state=42, solver='saga').fit(X1.T, y1)

print('Train Score: ', dl.score(X1.T, y1))
print('Test Score: ', dl.score(X2.T, y2))

w = dl.coef_
w = w.squeeze()

w0 = dl.intercept_

print('w0: ', w0)
print('w: ', w)

x1 = np.array([-2, 2])
x2 = -1 /w[1] * (w0 + w[0] * x1)

print('x1: ', x1)
print('x2: ', x2)

plt.plot(X1[0, y1==0], X1[1, y1==0], 'sb', X1[0, y1==1], X1[1, y1==1], 'sr', alpha=0.4)
plt.plot(X2[0, y2==0], X2[1, y2==0], '*b', X2[0, y2==1], X2[1, y2==1], '*r', ms=10)
plt.plot(x1, x2, '--k')
plt.axis('scaled')
plt.axis([-2, 2, -1.5, 1.5])
plt.grid(True)
```

Podemos concluir que os resultados são os mesmos.

Neste caso, devido à simplicidade deste modelo, não é necessário regularização.<br>
Isto é comprovado ao termos os melhores resultados quando o termo de regularização é 100, ou seja, muito pouco restrito, ou mesmo quando não há regularização.<br>

<b>A regularização deve ser utilizada quando o modelo tem a capacidade de entrar em sobreaprendizagem.</b><br>
Neste caso, devido à simplicidade, a duas dimensões, é basicamente um plano passado por uma não-linearidade, então não tem flexibilidade suficiente para entrar em sobreaprendizagem. Também podiamos usar mais termos como x1^2, x2^2, x1*x2. Ai sim, o modelo podia começar a entrar em sobreaprendizagem.<br>
Caso o modelo tivesse um elevado número de dimensões, existe maior facilidade para o modelo entrar em sobreaprendizagem.

### Classificação Binária: Breast Cancer Dataset

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
D = load_breast_cancer()
X = D.data
y = D.target
fN = D.feature_names
```

```python
X1, X2, y1, y2 = train_test_split(X, y, test_size=1/3)

dl = LogisticRegression(penalty='l2', C=1, random_state=42, solver='saga', max_iter=5000).fit(X1, y1)
```

Dividimos os dados em treino e teste e treinamos o discrminante.<br>
Para este caso, no discriminante logistico, adicionámos o parametro "max_iter". Isto porque se não fosse feito, o discriminante queixava-se que não estava a convergir.

```python
print('Train Score: ', dl.score(X1, y1))
print('Test Score: ', dl.score(X2, y2))
```

Curiosamente, o score do teste é melhor que o score do treino. Isto é muito incomum.

Vamos observar os coeficientes.

```python
w = dl.coef_
w = w.squeeze()

print(w)
```

Não é muito fácil analisar estes dados desta forma. Vamos usar o plot para termos uma melhor noção do que está a acontecer.

```python
plt.plot(w, '.')
plt.grid(True)
```

Podemos concluir que todos os pesos têm valores muito pequenos.<br>
Maior parte deles está muito próximo de 0, mas não são 0.

Vamos agora observar com a regularização l1.

```python
dl = LogisticRegression(penalty='l1', C=1, random_state=42, solver='saga', max_iter=5000).fit(X1, y1)

print('Train Score: ', dl.score(X1, y1))
print('Test Score: ', dl.score(X2, y2))

w = dl.coef_
w = w.squeeze()

print(w)

plt.plot(w, '.')
plt.grid(True)
```

Os scores continuam a ser muito incomuns, com o score do teste superior ao de treino.<br>
Agora podemos ver que os valores dos pesos já se encontram a 0. Vamos ver quantos são 0.

```python
print(np.sum(w == 0))
```

O que podemos concluir é que o classificador está a considerar que estas dimensões podem ser descartadas.

Vamos identificar as dimensões (features) que o determinante está a considerar como importantes para a classificação, assim como os respectivos indices.

```python
print(fN[w != 0])
print(np.argwhere(w != 0).squeeze())
```

Vamos agora repetir este processo, mas com o termo de regularização maior, ou seja, um C menor.

```python
dl = LogisticRegression(penalty='l1', C=0.1, random_state=42, solver='saga', max_iter=5000).fit(X1, y1)

print('Train Score: ', dl.score(X1, y1))
print('Test Score: ', dl.score(X2, y2))

w = dl.coef_
w = w.squeeze()

print(w)

plt.plot(w, '.')
plt.grid(True)
```

Podemos ver que existem menos dimensões diferentes de 0, embora o score se mantenha igual.

Vamos observar estes pesos com mais detalhe.

```python
print(np.sum(w == 0))
print(fN[w != 0])
print(np.argwhere(w != 0).squeeze())
```

Podemos concluir que 21 dimensões são descartadas, das 30 dimensões existentes.

### Classificação Multi-classe: Iris Dataset

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
D = load_iris()
X = D.data
y = D.target


X1, X2, y1, y2 = train_test_split(X, y, test_size=1/3, shuffle=True, random_state=0)
```

```python
plt.figure(figsize=(5, 8))

plt.plot(X1[y1==0, 0], X1[y1==0, 2], 'sr', X1[y1==1, 0], X1[y1==1, 2], 'sb', X1[y1==2, 0], X1[y1==2, 2], 'sg', alpha=0.4)
plt.plot(X2[y2==0, 0], X2[y2==0, 2], '*r', X2[y2==1, 0], X2[y2==1, 2], '*b', X2[y2==2, 0], X2[y2==2, 2], '*g', ms=10)
plt.axis('scaled')
plt.grid(True)
```

O train_test_split já baralha os dados automaticamente por pré-definição. Ainda assim, é melhor especificar pois podem haver updates e os métodos podem ser mudados.

Vamos classificá-los.

```python
dl = LogisticRegression(solver='liblinear', C=0.1, penalty='l1').fit(X1, y1)
y2e = dl.predict(X2)

print("Número de erros (teste): ", np.sum(y2 != y2e))
print(dl.coef_)
```

```python
dl = LogisticRegression(solver='liblinear', C=0.1, penalty='l2').fit(X1, y1)
y2e = dl.predict(X2)

print("Número de erros (teste): ", np.sum(y2 != y2e))
print(dl.coef_)
```

Como podemos observar, com a regularização l1, existem vários coeficientes a 0, mas na regularização l2 já não.<br>
Verificamos também que com a regularização l2 o número de erros diminui ligeiramente.

Vamos observar esses erros:

```python
plt.figure(figsize=(5, 8))

plt.plot(X1[y1==0, 0], X1[y1==0, 2], 'sr', X1[y1==1, 0], X1[y1==1, 2], 'sb', X1[y1==2, 0], X1[y1==2, 2], 'sg', alpha=0.4)
plt.plot(X2[y2==0, 0], X2[y2==0, 2], '*r', X2[y2==1, 0], X2[y2==1, 2], '*b', X2[y2==2, 0], X2[y2==2, 2], '*g', ms=10)
plt.plot(X2[y2!=y2e, 0], X2[y2!=y2e, 2], 'ok', alpha=0.5, ms=12)
plt.axis('scaled')
plt.grid(True)
```

Podemos observar que os erros são sobretudo da segunda classe.

Vamos ver agora "one over the rest" (ovr). Isto continua a ser um classificador binário, mas está a ser aplicado a um problema multi-classe.

```python
dl = LogisticRegression(solver='saga', C=0.1, penalty='l2', multi_class='ovr', max_iter=1000).fit(X1, y1)
y2e = dl.predict(X2)

print("Número de erros (teste): ", np.sum(y2 != y2e))
print(dl.coef_)
```

Como podemos observar, existem menos erros e podemos observar que não exsite nenhum parametro a 0, visto estar a ser utilizada a regularização l2.

Vamos ver a classificação multi-classe em vez do 'ovr'.

```python
dl = LogisticRegression(solver='saga', C=0.1, penalty='l2', multi_class='multinomial', max_iter=1000).fit(X1, y1)
y2e = dl.predict(X2)

print("Número de erros (teste): ", np.sum(y2 != y2e))
print(dl.coef_)
```

Assim, podemos concluir que este tipo de classificação é muito mais apropriado para este tipo de problemas, dado que apenas obtemos 3 erros, o menor número de erros até agora.

Vamos observar o 'ovr' para o tipo l1, assim como o 'multinomial'

```python
dl = LogisticRegression(solver='saga', C=0.1, penalty='l1', multi_class='ovr', max_iter=1000).fit(X1, y1)
y2e = dl.predict(X2)

print("Número de erros (teste): ", np.sum(y2 != y2e))
print(dl.coef_)
```

Embora o número de erros tenha reduzido, continuamos com mais erros.<br>
Também é de notar que apenas três pesos são diferentes de 0.

```python
dl = LogisticRegression(solver='saga', C=0.1, penalty='l1', multi_class='multinomial', max_iter=1000).fit(X1, y1)
y2e = dl.predict(X2)

print("Número de erros (teste): ", np.sum(y2 != y2e))
print(dl.coef_)
```

Este é o menor número possível de erros que obtemos. Apenas dois.
É possível obter bons resultados apenas com dois pesos, como podemos observar.

A conclusão é que vamos ter melhores resultados se especificarmos a classificação multi-classe num problema multi-classe, como seria de esperar.

## Máquina de Suporte Vectorial - SVM

As Máquinas de Suporte Vectorial são muito parecidas com os Discriminantes Lineares. Estas podem ser usadas tanto para a Regressão como para a Classificação. E até há algumas que são adaptadas a Clustering e a métodos não supervisionados.

Vamos ver SVM no contexto de classificação. Neste contexto, estas não são nada mais que classificadores binários.

### Exemplo de Dados Gaussianos

```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC, SVC
```

```python
Nc = 100
x1 = rd.randn(2, Nc) * 0.6 - 1
x2 = rd.randn(2, Nc) * 0.6 + 1
y = np.hstack((np.zeros(Nc), np.ones(Nc)))
X = np.hstack((x1, x2))
```

```python
plt.figure(figsize=(5, 7))
plt.plot(x1[0, :], x1[1, :], 'sb', mfc='w')
plt.plot(x2[0, :], x2[1, :], '^r', mfc='w')
plt.axis('scaled')
plt.grid(True)
```

Vamos definir dois modelos de máquinas de suporte vectorial

```python
svm1 = LinearSVC(dual=False, C=1).fit(X.T, y)
svm2 = SVC(kernel='linear', C=1).fit(X.T, y)

y1 = svm1.predict(X.T)
y2 = svm2.predict(X.T)
```

Para o LinearSVC, usamos o 'dual=False' pois número de dados é maior que o número de dimensões.

<b>Não esquecer que esta abordagem de treinar o modelo com todos os dados é errada.</b> Neste caso, apenas queremos observar comportamento da SVM e comparar a implementação do LinearSVC com o SVC.

Vamos analisar o número de erros em ambos os modelos.

```python
print(np.sum(y!=y1), np.sum(y!=y2))
```

Conseguimos observar, como expectável tendo em conta os plots, que os modelos não têm erros.

Vamos observar os pesos óptimos que ambos os modelos calcularam.

```python
w0a = svm1.intercept_
wa = svm1.coef_.squeeze()

w0b = svm2.intercept_
wb = svm2.coef_.squeeze()

print("LinearSVC")
print(w0a, wa)
print("---------------------------------------")
print("SVC")
print(w0b, wb)
```

Estes pesos são os que vão delimitar a reta que divide as classes dos dados, a chamada fronteira de decisão.<br>
Vamos mostrar essa reta.

```python
gx = np.array([-3, 3])
gy1 = -1 / wa[1] * (wa[0] * gx + w0a)
gy2 = -1 / wb[1] * (wb[0] * gx + w0b)

plt.figure(figsize=(5, 7))
plt.plot(x1[0, :], x1[1, :], 'sb', mfc='w')
plt.plot(x2[0, :], x2[1, :], '^r', mfc='w')
plt.plot(X[0, y!=y1], X[1, y!=y1], 'ok', ms=3)
plt.plot(gx, gy1, ':k')
plt.plot(gx, gy2, ':k')
plt.axis('scaled')
plt.axis([-3, 3, -3, 3])
plt.grid(True)
```

Embora nos vídeos do engenheiro existam erros e as fronteiras de decisão sejam as mesmas, não acontece isso neste exemplo.

Todas as retas entre aquelas duas, seriam fronteiras de decisão válidas. O modelo simplesmente calcula as que têm maior margem entre os pontos das classes.

Vamos repetir o processo, mas vamos calcular a margem de decisão manualmente.

```python
svm = LinearSVC(dual=False, C=1).fit(X.T, y)
ye = svm.predict(X.T)

print("Número de Erros: ", np.sum(y!=ye))

w0 = svm.intercept_
w = svm.coef_.squeeze()

print("Pesos: ", w0, w)

gx = np.array([-3, 3])
gy = -1 / w[1] * (w[0] * gx + w0)

mg = 1/np.sqrt(np.sum(w**2))
gy1 = -1 / w[1] * (w[0] * gx + w0 - mg)
gy2 = -1 / w[1] * (w[0] * gx + w0 + mg)

plt.figure(figsize=(5, 7))
plt.plot(x1[0, :], x1[1, :], 'sb', mfc='w')
plt.plot(x2[0, :], x2[1, :], '^r', mfc='w')
plt.plot(X[0, y!=ye], X[1, y!=ye], 'ok')
plt.plot(gx, gy, 'k')
plt.plot(gx, gy1, ':k')
plt.plot(gx, gy2, ':k')
plt.axis('scaled')
plt.axis([-3, 3, -3, 3])
plt.grid(True)
```

As linhas a tracejado são as margens deste modelo.<br>
Todos os valores que estejam entre a margem e a linha contínua, são os vectores de suporte do modelo.

Vamos alterar o valor de C e ver a sua influência para o mesmo caso.

```python
svm = LinearSVC(dual=False, C=0.1).fit(X.T, y)
ye = svm.predict(X.T)

print("Número de Erros: ", np.sum(y!=ye))

w0 = svm.intercept_
w = svm.coef_.squeeze()

print("Pesos: ", w0, w)

gx = np.array([-3, 3])
gy = -1 / w[1] * (w[0] * gx + w0)

mg = 1/np.sqrt(np.sum(w**2))
gy1 = -1 / w[1] * (w[0] * gx + w0 - mg)
gy2 = -1 / w[1] * (w[0] * gx + w0 + mg)

plt.figure(figsize=(5, 7))
plt.plot(x1[0, :], x1[1, :], 'sb', mfc='w')
plt.plot(x2[0, :], x2[1, :], '^r', mfc='w')
plt.plot(X[0, y!=ye], X[1, y!=ye], 'ok')
plt.plot(gx, gy, 'k')
plt.plot(gx, gy1, ':k')
plt.plot(gx, gy2, ':k')
plt.axis('scaled')
plt.axis([-3, 3, -3, 3])
plt.grid(True)
```

Como podemos concluir, quanto mais restrito for o termo de regularização, ou seja, quanto menor for o C, mais vectores de suporte este modelo vai ter.<br>
Quantos mais vectores de suporte o modelo tem, mais pesado fica o algoritmo. Isto porque o algoritmo só precisa dos vectores de suporte para classificar todos os dados.

Vamos agora aumentar o termo de regularização e observar as diferenças.

```python
svm = SVC(kernel='poly', degree=3, C=1, coef0=1, gamma='auto').fit(X, y)
ye = svm.predict(X)

vs = svm.support_vectors_
idx = svm.support_

print(idx)
```

Vamos visualizar estes vectores e as fronteiras de decisão.

```python
xx = np.linspace(-1.5, 2.5, 30)
yy = np.linspace(-1.5, 2.5, 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svm.decision_function(xy).reshape(XX.shape)

plt.figure(figsize=(10, 7))
plt.plot(X[y==0, 0], X[y==0, 1], 'sb', mfc='w')
plt.plot(X[y==1, 0], X[y==1, 1], '^r', mfc='w')

vs = svm.support_vectors_
idx = svm.support_

plt.plot(vs[y[idx]==0, 0], vs[y[idx]==0, 1], 'sb')
plt.plot(vs[y[idx]==1, 0], vs[y[idx]==1, 1], '^r')

print("Número de Vectores de Suporte: ", vs.shape[0])
print("Número de Erros: ", np.sum(y!=ye))
plt.plot(X[y!=ye, 0], X[y!=ye, 1], 'ok', ms=10, alpha=0.75)

ax = plt.gca()
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.axis('scaled')
plt.grid(True)
```

As primeiras 5 funções permitem criar uma grelha de pontos.<br>
O 'Z' obtém o resultado das funções de decisão da SVM.
É no 'ax.contour' que são criadas as margens e a fronteira. Isto vem do 'Z' ser passado como argumento e também possuir o 'levels' com três valores diferentes. -1 e 1 são as duas margens e o 0 é a fronteira.

Vamos observar o que acontece às fronteiras quando o termo diminuimos e aumentamos o termo de regularização.

## Ajuste de Parametros dos Modelos

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
D = load_breast_cancer()
X = D.data
y = D.target

X1, X2, y1, y2 = train_test_split(X, y, test_size=1/2, stratify=y)

sc = StandardScaler().fit(X1)
X1n = sc.transform(X1)
X2n = sc.transform(X2)

topScore = 0
gList = [0.0001, 0.001, 0.01, 0.1, 1]
cList = [0.1, 1, 10, 50, 100, 150, 200]

for c in cList:
    for g in gList:
        svm = SVC(kernel = 'rbf', gamma = g, C = c).fit(X1n, y1)
        score = svm.score(X2n, y2)
        if score >= topScore:
            topScore = score
            par = (g, c)
            
svm = SVC(kernel = 'rbf', gamma = par[0], C = par[1]).fit(X1n, y1)

print("% de acertos no treino: ", svm.score(X1n, y1))
print("% de acertos no teste: ", svm.score(X2n, y2))
print("Parametros: gamma=%.4f, C=%.1f: "%(par[0], par[1]))
```

Sempre que corrermos o código acima, vamos ter diferentes valores.<br>
Isto porque o train_test_split dá valores diferentes ao treino e teste do que lá estava previamente. Não é um erro.

O que torna este código um erro é o facto de inflacionarmos os parametros pois estamos a olhar para o conjunto de teste para saber se o resultado é bom. E isso não deve ser feito.

Assim, de forma a termos um grupo de dados onde podemos observar quais os melhores resultados, vamos criar um conjunto de validação.

#### Validação

```python
D = load_breast_cancer()
X = D.data
y = D.target

X1, X2, y1, y2 = train_test_split(X, y, test_size=1/3, stratify=y)

sc = StandardScaler().fit(X1)
X1n = sc.transform(X1)
X2n = sc.transform(X2)

X1na, X1nb, y1a, y1b = train_test_split(X1n, y1, test_size=1/2, stratify=y1)

topScore = 0
gList = [0.0001, 0.001, 0.01, 0.1, 1]
cList = [0.1, 1, 10, 50, 100, 150, 200]

for c in cList:
    for g in gList:
        svm = SVC(kernel = 'rbf', gamma = g, C = c).fit(X1na, y1a)
        score = svm.score(X1nb, y1b)
        if score >= topScore:
            topScore = score
            par = (g, c)
            
svm = SVC(kernel = 'rbf', gamma = par[0], C = par[1]).fit(X1n, y1)

print("% de acertos no treino: ", svm.score(X1n, y1))
print("% de acertos na validação: ", topScore)
print("% de acertos no teste: ", svm.score(X2n, y2))
print("Parametros: gamma=%.4f, C=%.1f: "%(par[0], par[1]))
```

Agora sim, as coisas estão feitas de forma correcta.<br>
Temos três grupos: Treino, Validação e Teste.<br>
Treinamos o modelo com os dados de treino, verificamos o score para os dados de validação e, caso estes sejam os melhores, guardamo-los.Treinamos novamente o modelo, mas desta vez com todo o conjunto de treino, treino + validação, e testamos com os dados de teste.

#### Validação Cruzada

```python
from sklearn.model_selection import cross_val_score
```

```python
D = load_breast_cancer()
X = D.data
y = D.target

X1, X2, y1, y2 = train_test_split(X, y, test_size=1/4, stratify=y)

sc = StandardScaler().fit(X1)
X1n = sc.transform(X1)
X2n = sc.transform(X2)

topScore = 0
gList = [0.0001, 0.001, 0.01, 0.1, 1]
cList = [0.1, 1, 10, 50, 100, 150, 200]

for c in cList:
    for g in gList:
        svm = SVC(kernel = 'rbf', gamma = g, C = c)
        scores = cross_val_score(svm, X1n, y1, cv=5)
        if np.mean(scores) >= topScore:
            topScore = np.mean(scores)
            par = (g, c)
            
svm = SVC(kernel = 'rbf', gamma = par[0], C = par[1]).fit(X1n, y1)

print("% de acertos no treino: ", svm.score(X1n, y1))
print("% de acertos na validação: ", topScore)
print("% de acertos no teste: ", svm.score(X2n, y2))
print("Parametros: gamma=%.4f, C=%.1f: "%(par[0], par[1]))
```

Esta mesmo forma de ajustar os parâmetros já está implementada no sklearn.

```python
D = load_breast_cancer()
X = D.data
y = D.target

X1, X2, y1, y2 = train_test_split(X, y, test_size=1/4, stratify=y)

sc = StandardScaler().fit(X1)
X1n = sc.transform(X1)
X2n = sc.transform(X2)

topScore = 0
gList = [0.0001, 0.001, 0.01, 0.1, 1]
cList = [0.1, 1, 10, 50, 100, 150, 200]
grelha = {'gamma' : gList, 'C' : cList}

gSearch=GridSearchCV(SVC(kernel = 'rbf'), param_grid= grelha, cv=5)
gSearch.fit(X1n, y1)

svm = gSearch.best_estimator_
par = gSearch.best_params_

print("% de acertos no treino: ", svm.score(X1n, y1))
print("% de acertos no teste: ", svm.score(X2n, y2))
print("Parametros: gamma=%.4f, C=%.1f: "%(par['gamma'], par['C']))
```
