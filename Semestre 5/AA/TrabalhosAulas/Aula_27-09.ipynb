```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rd
import scipy.linalg as la
```

```python
A = np.array([[1/3, 1], [-1/3, 1]])/np.sqrt(2)
A
```

```python
X = rd.randn(2, 10000)
b = np.array([3, 2])
X2 = np.dot(A, X) + b[:, np.newaxis]
plt.plot(X[0, :], X[1, :], '.', X2[0, :], X2[1, :], '.')
plt.axis('scaled')
plt.grid(True)
```

```python
c = np.array([-20, 10])
X3 = X2+c[:, np.newaxis]
plt.plot(X[0, :], X[1, :], '.', X2[0, :], X2[1, :], '.', X3[0, :], X3[1, :], '.')
plt.axis('scaled')
plt.grid(True)
```

```python
C2 = np.cov(X2)
C2
```

```python
C3 = np.cov(X3)
C3
```

```python
#A média variar não altera os valores de covariância.
#A ideia é: o quanto é que os pontos variam à volta da média.
```
