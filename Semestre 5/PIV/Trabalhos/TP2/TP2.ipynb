```python
import numpy as np
import cv2
from bs4 import BeautifulSoup
import os
```

```python
#### Métodos Auxiliares
def showImage(title, image):
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def loadAllImages(filePath):
    parque_NomesFicheiros = os.listdir(filePath)
    parque_NumeroImagens = int(len(parque_NomesFicheiros) / 2)
    parque_Imagens = [None] * parque_NumeroImagens

    index = 0
    for nomeFicheiro in parque_NomesFicheiros:
        if nomeFicheiro[-4:] == ".jpg":
            parque_Imagens[index] = cv2.imread((filePath + "/" + nomeFicheiro))
            index = index + 1
    
    return parque_Imagens
```

## Parque 2012-09-12

```python
data = open('Imagens/2012-09-12/2012-09-12_06_20_57.xml', 'r').read()
Bs_data = BeautifulSoup(data, "xml")
parkingSpacesData = Bs_data.find_all('space')

parque1 = cv2.imread("Imagens/2012-09-12/2012-09-12_06_20_57.jpg")
boxes = [None] * len(parkingSpacesData)

#print(parkingSpacesData[0].contents[1].contents[1].get('x'))
# print(parkingSpacesData[0].contents[1].contents[3].get('w'))
# print(parkingSpacesData[0].contents[1].contents[5].get('d'))

for index in range(len(boxes)):
    centerData = parkingSpacesData[index].contents[1].contents[1]
    center = (int(centerData.get('x')), int(centerData.get('y')))
    
    sizeData = parkingSpacesData[index].contents[1].contents[3]
    size = (int(sizeData.get('w')), int(sizeData.get('h')))
    
    rotationData = parkingSpacesData[index].contents[1].contents[5]
    rotation = int(rotationData.get('d'))
    
    rotationRectangle = (center, size, rotation)
    box = cv2.boxPoints(rotationRectangle)
    box = np.int0(box)
    boxes[index] = box

    rectangle = cv2.drawContours(parque1,[box],0,(255,255,255),2)
    
showImage("P1", parque1)
```
