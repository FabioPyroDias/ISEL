```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
```

```python
n = np.arange(1000)
x = 10 + 2 * np.cos(np.pi / 6 * n) + 10 * np.cos(np.pi / 3 * n)
```

```python
#Filtro Passa Baixo
nCoeficientesPoucoQualidade = 11
nCoeficientesMuitaQualidade = 101
freqCorte = np.pi / 8
bk = ss.firwin(nCoeficientesMuitaQualidade, freqCorte, pass_zero='lowpass', fs=2*np.pi)

#Resposta em Frequência
w, H = ss.freqz(bk)

#Filtragem - Implementa a equação às diferenças
y = ss.lfilter(bk, 1.0, x)
```

```python
plt.plot(w, np.abs(H))
```

```python
plt.plot(n, x)
plt.plot(n, y)
```

```python
#Filtro Passa Banda
nCoeficientesPoucoQualidade = 11
nCoeficientesMuitaQualidade = 101
nCoeficientesExtremaQualidade = 201
freqCorte1 = np.pi / 8
freqCorte2 = np.pi / 4
bk = ss.firwin(nCoeficientesExtremaQualidade, [freqCorte1, freqCorte2], pass_zero='bandpass', fs=2*np.pi)

#Resposta em Frequência
w, H = ss.freqz(bk)

#Filtragem - Implementa a equação às diferenças
y = ss.lfilter(bk, 1.0, x)
```

```python
plt.plot(w, np.abs(H))
```

```python
plt.plot(n, x)
plt.plot(n, y)
```

```python
def FiltroComposto(x):
    # Filtro Passa Baixo
    nCoeficientesExtremaQualidade = 201
    freqCorte = np.pi / 5
    bk1 = ss.firwin(nCoeficientesExtremaQualidade, freqCorte, pass_zero='lowpass', fs=2*np.pi)
    bk1 = 2 * bk1

    #Resposta em Frequência
    #w, H = ss.freqz(bk)

    #Filtragem - Implementa a equação às diferenças
    #y1 = ss.lfilter(bk, 1.0, x)

    #Filtro Passa Banda
    nCoeficientesExtremaQualidade = 201
    freqCorte1 = np.pi / 5
    freqCorte2 = np.pi / 4
    bk2 = ss.firwin(nCoeficientesExtremaQualidade, [freqCorte1, freqCorte2], pass_zero='bandpass', fs=2*np.pi)

    #Resposta em Frequência
    #w, H = ss.freqz(bk)

    #Filtragem - Implementa a equação às diferenças
    #y2 = ss.lfilter(bk, 1.0, x)

    #Filtro Passa Alto
    nCoeficientesExtremaQualidade = 201
    freqCorte = np.pi / 2
    bk3 = ss.firwin(nCoeficientesExtremaQualidade, freqCorte, pass_zero='highpass', fs=2*np.pi)
    bk3 = 5*bk3

    #Resposta em Frequência
    #w, H = ss.freqz(bk)

    #Filtragem - Implementa a equação às diferenças
    #y3 = ss.lfilter(bk, 1.0, x)

    #Filtro Final
    bk = bk1 + bk2 + bk3
    
    #Resposta em Frequência
    w, H = ss.freqz(bk)
    
    #Filtragem
    yTotal = ss.lfilter(bk, 1.0, x)
    
    return yTotal, w, H
```

```python
n = np.arange(1000)

y, w, H = FiltroComposto(x)

plt.plot(w, np.abs(H))
plt.show()
plt.plot(n, x)
plt.plot(n, y)
plt.show()
```
