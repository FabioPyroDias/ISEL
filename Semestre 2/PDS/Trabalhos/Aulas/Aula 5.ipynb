$x(t) = 2 + \cos(2 \pi 4t - \pi / 4)$

$y(t) = x(t+0.1) + \cos(2 \pi 6t)$

Usando a equação de Síntese da Série de Fourier:

$y(t) = \sum_{k=-\infty}^{\infty} Y_k e^{j 2 \pi f_0 k t}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 100
t = np.arange(0, 1, 1/Fs)

y = 2 + 4 * np.cos(2 * np.pi * 4 * (t + 0.1) - np.pi / 4) + np.cos(2 * np.pi * 6 * t)

plt.plot(t, y)

#Hardcoded
#y = 2 + 2*np.exp(2j*np.pi*4*t) * np.exp(-1j*np.pi/4) + 2*np.exp(-2j*np.pi*4*t) * np.exp(1j*np.pi/4) + 0.5*np.exp(2j*np.pi*6*t) * np.exp(0) + 0.5*np.exp(-2j*np.pi*6*t) * np.exp(0)
#plt.plot(t, y)

fase = 2*np.pi*0.4 - np.pi/4

kIndex = np.array([-3, -2, 0, 2, 3])
#XkIndex = np.array([0.5, 2*np.exp(1j*np.pi/4), 2, 2*np.exp(-1j*np.pi/4), 0.5])
XkIndex = np.array([0.5, 2*np.exp(-1j*fase), 2, 2*np.exp(1j*fase), 0.5])

#kIndex = np.array([-1, 0, 1])
#XkIndex = np.array([2*np.exp(1j*np.pi/4), 2, 2*np.exp(-1j*np.pi/4)])

eqSinteseSerieFourier = np.zeros(len(t))

f0 = 2

for index in range(len(kIndex)):
    eqSinteseSerieFourier = eqSinteseSerieFourier + (XkIndex[index] * np.exp(2j*np.pi*kIndex[index]*f0*t))

plt.plot(t, eqSinteseSerieFourier, 'r')
```
