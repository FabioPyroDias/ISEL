```python
import numpy as np
```

```python
def sistema(x):
    
    y = np.zeros(len(x))
    
    for i in range(len(x)):
        
            if(i == 0):
                y[i] = x[0]
            else:
                y[i] = x[i] - x[i - 1]
        
    return y
```

```python
x = [1, 2, 3, 4, 5]
y = sistema(x)
y
```

```python
h1 = [1, -1]
y1conv = np.convolve(x, h1)
y1conv
```

```python
def sistemaDelay(x, K):
    y = np.zeros(len(x))
        
    for i in range(len(x)):
        if i < K:
            y[i] = x[i]
        else:
            y[i] = x[i] + x[i - K]
    
    return y
```

```python
xDelay = [1, 1, 2, 3, 4, 2, 4, 5, 5, 7, 8, 2, 1, 4, 2, 4, 6, 3, 5, 3, 5]
yDelay = sistemaDelay(xDelay, 3)
yDelay
```

```python
import matplotlib.pyplot as plt
```

```python
w = np.arange(0, np.pi, 0.001)
Hw = 0.5 + 0.5*np.exp(-1j*w)
Hw
```

```python
plt.plot(w, np.abs(Hw))
plt.figure()
plt.plot(w, np.angle(Hw))
```

```python
import scipy.signal as ss
```

```python
w, h = ss.freqz([0.5, 0.5], 1.0)
plt.plot(w, np.abs(h))
```

```python
n = np.arange(1000)
x = 10 + 2 * np.cos(np.pi / 6 * n) + 10 * np.cos(np.pi / 3 * n)
bk = [1.0, -1.0]
y1filter = ss.lfilter(bk, [1.0], x)
```

```python
plt.plot(n, x)
plt.plot(n, y1filter)
```
