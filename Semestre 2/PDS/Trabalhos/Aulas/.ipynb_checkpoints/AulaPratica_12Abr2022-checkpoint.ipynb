```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(-5,5,0.001)
u = t>=0
x = 4*np.cos(2*np.pi*50*t)
y = x*u
```

```python
u*1.
u2 = t-2>=0
plt.plot(t,u*1.-u2)
```

```python
plt.plot(t,x*(u*1.-u2))
```

```python
Fs = 1000
duracao = 1
t = np.arange(0,duracao,1/Fs)
fnota=440
nota = np.cos(2*np.pi*fnota*t)
m = -1/duracao
b = 1 
envelope = m*t+b
plt.plot(t,envelope)
```

```python
y = nota * envelope
plt.plot(t,y)
```
