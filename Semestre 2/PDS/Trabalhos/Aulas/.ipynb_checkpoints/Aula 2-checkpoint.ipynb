```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 1, 0.0001)

x = 10 * np.cos(2 * np.pi * 440 * t)
y = 10 * np.cos(2 * np.pi * 440 * t + 0.4 * np.pi)

plt.plot(t, x, 'black')
plt.plot(t, y, 'o-')
plt.grid()
plt.xlabel('t(s)')
plt.legend(['x', 'y'])
```

```python
from IPython.display import Audio, display
display(Audio(data = x, rate = 1 / 0.0001))
```
