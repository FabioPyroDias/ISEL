```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display
```

```python
Fs = 8000
Fc = 400
Fdelta = 10

t = np.arange(0, 1, 1/Fs)

f1 = Fc - Fdelta
f2 = Fc + Fdelta

x1 = 9 * np.cos(2 * np.pi * f1 * t)
x2 = 9 * np.cos(2 * np.pi * f2 * t)
y = x1 * x2

plt.plot(t, y)
display(Audio(data=y, rate=Fs))
```
